```python
import csv
import requests
from bs4 import BeautifulSoup  # 从bs4引入BeautifulSoup

# 请求网页
url = "https://book.douban.com/tag/%E6%91%84%E5%BD%B1?start=0&type=T"
response = requests.get(url)
# 初始化BeautifulSoup方法一：利用网页字符串自带的编码信息解析网页
soup = BeautifulSoup(response.content.decode('utf-8'), 'lxml')

# 初始化BeautifulSoup方法二：手动指定解析编码解析网页
# soup = BeautifulSoup(response.content, 'lxml', from_encoding='utf-8')

#print(soup)  # 输出BeautifulSoup转换后的内容
books = soup.find('li', id="subject-list")  # 先找到最大的div
#print(books)  # 输出最大的div的内容

csv_file = open('book_rank.csv', 'w', encoding="utf-8", newline='')
writer = csv.writer(csv_file)

writer.writerow(["书名","链接", "信息", "评分", "摘要"])  # 写入标题
for each_book in soup.find_all('li', class_="subject-item"):  # 从最大的div里面找到影片的div
    # print(each_movie)  # 输出每个影片div的内容
    all_h2_tag = each_book.find_all('h2')
    all_a_tag = each_book.find_all('a')
    all_div_tag = each_book.find_all('div')
    all_span_tag = each_book.find_all('span')
    all_p_tag = each_book.find_all('p')
    book_name = all_h2_tag[0].text
    book_href = all_a_tag[1]['href']
    book_info = all_div_tag[2].text
    book_rating = all_span_tag[1].text+all_span_tag[2].text
    book_excerpt = all_p_tag[0].text
    print('书名：{}，链接：{}，信息：{}，评分：{}，摘要：{}'.format(
        book_name, book_href, book_info, book_rating, book_excerpt))
    writer.writerow([book_name, book_href, book_info, book_rating, book_excerpt])

csv_file.close()
print("write_finished!")
```
