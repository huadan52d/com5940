```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from random import randint
from time import time
```

```python
header = ['标题','链接','信息']
data = []
```

```python
for i in range(0,5):
    quote_page = requests.get('https://book.douban.com/tag/%E6%91%84%E5%BD%B1?start='+str(i*20))
    sleep(randint(8,15))
    print("\n***** Page " + str(i+1) +" in action *****")
    soup = BeautifulSoup(quote_page.content,'html.parser')
    
    for article in soup.find_all("div", attrs={"class":"info"}):
        page_no = str(i+1)
        title = article.a.text.encode('utf-8').strip()
        decoded_title = title.decode('utf-8')        
        print(decoded_title)
        url = article.a.get('href')
        print(url)
        all_div_tag = article.find_all('div')
        info = all_div_tag[0].text.encode('utf-8').strip()
        decoded_info = info.decode('utf-8')
        print(decoded_info)
        data.append((decoded_title,url,decoded_info))
        
df = pd.DataFrame(data,
    columns = header
)
df.to_csv('db-book.csv', sep='\t', encoding='utf-8')
```
